```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.optim as optim
from torchvision import transforms,models
```

```python
##loading the pretrained model features
vgg19 = models.vgg19(pretrained=True).features
```

```python
##freezing the parameters
for param in vgg19.parameters():
    param.requires_grad_(False)
    
    
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(vgg19.to(device))
```

```python
def load_image(img_path, max_size=400, shape = None):
    '''
    img_path: loads the path of an image
    image_size: limitation if the image size is greater than the maxsize
    returns the image with tranformations by sqeezing in RGB'''
    img = Image.open(img_path).convert('RGB')
    
    if max(img.size) > max_size:
        size = max_size
    else:
        size = max(img.size)
        
    if shape is not None:
        size= shape
        
        
    transform = transforms.Compose([transforms.Resize(size),
                                   transforms.ToTensor(),
                                   transforms.Normalize((0.485, 0.456, 0.406), 
                                             (0.229, 0.224, 0.225))])
    
    img = transform(img)[:3, :, :].unsqueeze(0)
    
    return img
```

```python
##loading the content Image
content_img = load_image('E:/Deep learning/Style_Transfer/content.jpg').to(device)

##loading the Style Image
style_img = load_image('E:/Deep learning/Style_Transfer/wave.jpg', shape=content_img.shape[-2:]).to(device)
```

```python
def im_convert(tensor):
    """ Display a tensor as an image. """
    
    image = tensor.to("cpu").clone().detach()
    
    image = image.numpy().squeeze()
    image = image.transpose(1,2,0)
    image = image * np.array((0.229, 0.224, 0.225)) + np.array((0.485, 0.456, 0.406))
    image = image.clip(0, 1)

    return image
```

```python
fig, (fig1, fig2) = plt.subplots(1,2,figsize = (20,10))

fig1.imshow(im_convert(content_img))

fig2.imshow(im_convert(style_img))

fig1.title.set_text('content_img')
fig2.title.set_text('style_img')
```

```python
##calculating the GRAM_MATRIX###

def gram_matrix(tensor):
    
    _, d, h, w = tensor.size()
    
    tensor = tensor.view(d, h * w)
    
    gram = torch.mm(tensor, tensor.t())
    
    return gram
```

```python
def get_features(image, models, layers = None):
    '''layers frm VGG 16 pretrained model where the below
    layers are useful for content representaion of an image''' 
    
    
    if layers is None:
        layers = {'0': 'conv1_1',
                 '5' : 'conv2_1',
                  '10': 'conv3_1',
                  '19' : 'conv4_1',
                  '21' : 'conv4_2',    ###This is the Content Representation Layer##
                  '28' : 'conv5_1'
                  
                 }
        
        
    features = {}
    x = image
    
    for name, layer in models._modules.items():
        x = layer(x)
        if name in layers:
            features[layers[name]] = x
            
    return features
```

```python
content_features = get_features(content_img, vgg19)
style_features = get_features(style_img, vgg19)

style_grams = {}
for layers in style_features:
    style_grams[layers] = gram_matrix(style_features[layers])
    

    
###copyting the content_image which is used by changing iteratively for target image####
target = content_img.clone().requires_grad_(True).to(device)

print(style_grams)
```

```python
####weights for each layer in style represenattion i.e., conv1-1, conv2-1, conv3-1, conv4-1, conv5-1 b/w 0 and 1 ####

style_weights = {'conv1_1': 1.,
                 'conv2_1': 0.8,
                 'conv3_1': 0.5,
                 'conv4_1': 0.3,
                 'conv5_1': 0.1}

# you may choose to leave these as is
content_weight = 1  # alpha
style_weight = 1e3  # beta
```

```python
# for displaying the target image, intermittently
show_every = 400

# iteration hyperparameters
optimizer = optim.Adam([target], lr=0.003)
steps = 2000  # decide how many iterations to update your image (5000)

for ii in range(1, steps+1):
  
     
    target_features = get_features(target, vgg19)
    content_loss = torch.mean((target_features['conv4_2'] - content_features['conv4_2'])**2)
    
    # the style loss
    # initialize the style loss to 0
    style_loss = 0
    # iterate through each style layer and add to the style loss
    for layer in style_weights:
        # get the "target" style representation for the layer
        target_feature = target_features[layer]
        _, d, h, w = target_feature.shape
        
        
        target_gram = gram_matrix(target_feature)
        
        ## TODO:  get the "style" style representation
        style_gram = style_grams[layer]
        ## TODO: Calculate the style loss for one layer, weighted appropriately
        layer_style_loss = style_weights[layer] * torch.mean((target_gram - style_gram)**2)
        
        # add to the style loss
        style_loss += layer_style_loss / (d * h * w)
    total_loss = content_weight * content_loss + style_weight * style_loss
    
   
    # update your target image
    optimizer.zero_grad()
    total_loss.backward()
    optimizer.step()
    

    if  ii % show_every == 0:
        print('Total loss: ', total_loss.item())
        plt.imshow(im_convert(target))
        plt.show()
```

```python
fig, (fig1, fig2) = plt.subplots(1,2, figsize = (20,10))
fig1.imshow(im_convert(content_img))
fig2.imshow(im_convert(target))
```
